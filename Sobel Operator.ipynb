## Vertical and Horizontal filter

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
## Loading our image
img1=cv2.imread("diag7.jpg")
img=cv2.imread("diag7.jpg")
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
img1=cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
```

```python
kernel_ver= np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
kernel_diag=  np.array([[0,-1,-2],[1,0,-1],[2,1,0]])

kernel_horizontal=np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
print("Vertical Sobel operator\n\n",kernel_ver,"\n\nntal horizontal sobel operator\n\n",kernel_horizontal)

# def sobel_op(image, kernel):img_new=       
#     return img_new
img_diag=cv2.filter2D(img,-1,kernel_diag)
img_ver= cv2.filter2D(img,-1,kernel_ver)
img_hor= cv2.filter2D(img,-1,kernel_horizontal)
kernel_diag
```

```python
fig= plt.figure(figsize=(20,10))
fig.add_subplot(221),plt.imshow(img,cmap='gray'),plt.title("Original image")
fig.add_subplot(222),plt.imshow(img_hor,cmap='gray'),plt.title("filtered image by horizontal sobel operator")
fig.add_subplot(223),plt.imshow(img_ver,cmap='gray'),plt.title("filtered image by vertical sobel operator")
```
